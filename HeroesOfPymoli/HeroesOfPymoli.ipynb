```python
# initial set up
import pandas as pd

# csv file, my file is one level up
csvfile = "../Resources/purchase_data.csv"

# reading file and storing into panda data frame
purchase_data = pd.read_csv(csvfile)

# print and check headers
purchase_data.head()
```

```python
# Players count, total
total_players = len(purchase_data["SN"].unique())

total_df = pd.DataFrame({"Total Players": [total_players]})
total_df
```

```python
# Purchasing Analysis (Total)
# Run basic calculations to obtain number of unique items
unique_items = len(purchase_data["Item ID"].unique())
average_price = purchase_data["Price"].mean()
numb_purchases = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].sum()
# purchase_data.dtypes

# Create a summary data frame to hold the results
summary_df = pd.DataFrame ({
    "Number of Unique Items": [unique_items],
    "Average Price": [average_price],
    "Number of Purchases": [numb_purchases],
    "Total Revenue": [total_revenue]
})

# Optional: formatting to for better view of $
summary_df["Average Price"] = summary_df["Average Price"].map("${:,.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:,.2f}".format)

# display the summary data frame
summary_df
```

```python
# Gender Demographics
# Percentage and Count of Male Players # Percentage and Count of Female Players # Percentage and Count of Other/Non-Disclosed
# use GroupBy to separate the values, 04-Pandas/02-Activities/06-Ins_Groupby
grouped_gender_df = purchase_data.groupby(["Gender"])
# grouped_gender_df.count().head(), https://www.geeksforgeeks.org/python-pandas-series-nunique/
gender_count = grouped_gender_df["SN"].nunique()
gender_percent = gender_count / total_players * 100

# create a new summary table but doesn't say explicitly...weird
gender_demos_df = pd.DataFrame({
    "Total Count": gender_count, 
    "Percentage of Players": gender_percent.map("{:.2f}%".format)
})

# resetting the groupby table formatting - Gender moved to header line
gender_demos_df = gender_demos_df.reset_index()

# print summary table
gender_demos_df
```

```python
# Purchasing Analysis (Gender)
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
purchase_count = grouped_gender_df["Purchase ID"].count()
avg_purchase_price = grouped_gender_df["Price"].mean()
total_purchases = grouped_gender_df["Price"].sum()
avg_purchase_total_person = total_purchases / gender_count

# Create a summary data frame to hold the results & Optional: give the displayed data cleaner formatting
purchasing_summary_df = pd.DataFrame({
    "Purchase Count" : purchase_count,
    "Average Purchase Price" : avg_purchase_price.map("${:,.2f}".format),
    "Total Purchase Value" : total_purchases.map("${:,.2f}".format),
    "Ave Total Purchase per Person" : avg_purchase_total_person.map("${:,.2f}".format)
})

# resetting the groupby table formatting - Gender moved to header line
purchasing_summary_df = purchasing_summary_df.reset_index()

# Display the summary data frame
purchasing_summary_df
```

```python
# Age Demographics
# Establish bins for ages-- <10, 10-14, 15-19, 20-24, 25-29, 30-34, 35-39, 40+
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
age_label = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Categorize the existing players using the age bins. Hint: use pd.cut() 
# 03-Act/03-Ins-Binning df["Test Score Summary"] = pd.cut(df["Test Score"], bins, labels=group_names, include_lowest=True)
purchase_data["Ages"] = pd.cut(purchase_data["Age"], bins, labels=age_label, include_lowest=True)
grouped_ages_df = purchase_data.groupby(["Ages"])

#repeat my steps above to # Calculate the numbers and percentages by age group
age_count = grouped_ages_df["SN"].nunique()
ages_percent = age_count / total_players * 100

# Create a summary data frame to hold the results & # Optional: round the percentage column to two decimal points
age_demo_df = pd.DataFrame({
    "Total Count per Group" : age_count,
    "Percentage of Players" : ages_percent.map("{:.2f}%".format)
})

# resetting the groupby table formatting - Ages moved to header line
age_demo_df = age_demo_df.reset_index()

# Display Age Demographics Table
age_demo_df
```

```python
# Purchasing Analysis (Age)
# Bin the purchase_data data frame by age
# used the previous dataframe & copy/pasted previous code, changed names when needed
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
purchase_count_age = grouped_ages_df["Purchase ID"].count()
avg_purchase_age = grouped_ages_df["Price"].mean()
total_purchase_age = grouped_ages_df["Price"].sum()
avg_purchase_total_age = total_purchase_age / age_count

# Create a summary data frame to hold the results & # Optional: give the displayed data cleaner formatting
ages_summary_df = pd.DataFrame({
    "Purchase Count" : purchase_count_age,
    "Average Purchase Price" : avg_purchase_age.map("${:,.2f}".format),
    "Total Purchase Value" : total_purchase_age.map("${:,.2f}".format),
    "Ave Total Purchase per Person" : avg_purchase_total_age.map("${:,.2f}".format)
})

# ages_summary_df["Average Purchase Price"] = ages_summary_df["Average Purchase Price"].map("${:,.2f}".format)
# ages_summary_df["Total Purchase Value"] = ages_summary_df["Total Purchase Value"].map("${:,.2f}".format)
# ages_summary_df["Ave Total Purchase per Person"] = ages_summary_df["Ave Total Purchase per Person"].map("${:,.2f}".format)

# resetting the groupby table formatting - Gender moved to header line
ages_summary_df = ages_summary_df.reset_index()

# Display the summary data frame
ages_summary_df
```

```python
# Top Spenders
# Run basic calculations to obtain the results in the table below SN, Purchase count, Av Purch Price, Tot Purch Price
grouped_spenders_df = purchase_data.groupby(["SN"])

purchase_count_spend = grouped_spenders_df["Purchase ID"].count()
avg_purchase_spend = grouped_spenders_df["Price"].mean()
total_purchase_spend = grouped_spenders_df["Price"].sum()

# Create a summary data frame to hold the results & # Optional: give the displayed data cleaner formatting
spenders_summary_df = pd.DataFrame ({
    "Purchase Count" : purchase_count_spend,
    "Average Purchase Price" : avg_purchase_spend.map("${:,.2f}".format),
    "Total Purchase Value" : total_purchase_spend.map("${:,.2f}".format)
})
# Sort the total purchase value column in descending order, 
# 02-Act/07-Par_Pokedex strongest_pokemon_df = pokemon_comparison_df.sort_values(["Total"], ascending=False)
# spenders_summary_df["Average Purchase Price"] = spenders_summary_df["Average Purchase Price"].map("${:,.2f}".format)
# spenders_summary_df["Total Purchase Value"] = spenders_summary_df["Total Purchase Value"].map("${:,.2f}".format)
spenders_summary_df = spenders_summary_df.sort_values(["Purchase Count"], ascending=False)
spenders_summary_df = spenders_summary_df.reset_index()

# Display a preview of the summary data frame
spenders_summary_df.head(10)
```

```python
# Most Popular Items
# Retrieve the Item ID, Item Name, and Item Price columns
# create new data frame to house the above three items-duh Item Price is Price in the original df
pop_items_df = purchase_data[["Item ID", "Item Name", "Price"]]
# pop_items_df

# Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value
# grouped_spenders_df = purchase_data.groupby(["SN"])
grouped_item_df = pop_items_df.groupby(["Item ID","Item Name"])

purchase_count_pop = grouped_item_df["Item ID"].count()
avg_item_price = grouped_item_df["Price"].mean()
total_purchase_value = grouped_item_df["Price"].sum()


# Create a summary data frame to hold the results & # Optional: give the displayed data cleaner formatting
mostpop_summary_df = pd.DataFrame ({
    "Purchase Count" : purchase_count_pop,
    "Item Price" : avg_item_price.map("${:,.2f}".format),
    "Total Purchase Value" : total_purchase_value.map("${:,.2f}".format)
})

# Sort the purchase count column in descending order
# mostpop_summary_df["Total Purchase Value"] = mostpop_summary_df["Total Purchase Value"].map("${:,.2f}".format)
mostpop_summary_df = mostpop_summary_df.sort_values(["Purchase Count"], ascending=False)
mostpop_summary_df = mostpop_summary_df.reset_index()

# Display a preview of the summary data frame
mostpop_summary_df.head(10)
```

```python
# Most Profitable Items
# Sort the above table by total purchase value in descending order
# Optional: give the displayed data cleaner formatting

# had to use reset because the string couldn't be changed to a float without removing the $
profit_items_df = mostpop_summary_df.reset_index()
profit_items_df['tprofit'] = profit_items_df['Total Purchase Value'].str.lstrip('$')
profit_items_df.tprofit = profit_items_df.tprofit.astype(float)

# Display a preview of the data frame
profit_items_df.sort_values(["tprofit"], ascending = False).drop('tprofit', axis=1)

```
